Extract student data from excel to python dictionary

```python
# For Student data
df2 = pd.read_excel(r'Students.xlsx')
students = {}
student_names = df2['students'].values
student_topics = df2['topics'].values
empty = []
for i in student_topics:
    empty.append(i.split(', '))
student_topics = empty

for i in range(len(df2)):
    students[student_names[i]] = student_topics[i]
students
```

Construct graphs and find optimal solutions

```python
MAX_GROUP_SIZE = 3
def get_student_topic_supervisor_assignments(all_topics, students, supervisors, num_topics_used, max_group_size=MAX_GROUP_SIZE, do_supervisor_load_balancing=False):

    G = nx.DiGraph()
    G.add_node('sink', demand=len(students) - num_topics_used)
    %matplotlib inline
    # Add student and topic nodes and form edges
    for topic in all_topics:
        G.add_node(topic)
        G.add_edge(topic, 'sink', weight=0, capacity=max_group_size-1)

    for student in students:
        prefs = students[student]
        G.add_node(student, demand=-1)
        # add increasing weight edges from student to preferences (lowest == best)
        for i, topic in enumerate(prefs):
            G.add_edge(student, topic, weight=i, capacity=1)

    G.add_node('sink_2', demand=num_topics_used)

    # Add supervisor and topic nodes and form edges
    for topic in all_topics:
        G.add_node(topic + "_2")
        G.add_edge(topic, topic + "_2", weight=0, capacity=1)

    for supervisor in supervisors:
        supervisor_properties = supervisors[supervisor]
        for topic in supervisor_properties['topics']:
            G.add_edge(topic + "_2", supervisor, weight=0, capacity=1)
        if do_supervisor_load_balancing:
            for i in range(supervisor_properties['num_groups']):
                G.add_node(supervisor + "_dummy")
                G.add_edge(supervisor, supervisor +
                           "_dummy", weight=i, capacity=1)
                G.add_edge(supervisor + "_dummy",
                           'sink_2', weight=0, capacity=1)
        else:
            G.add_edge(supervisor, 'sink_2', weight=0,
                       capacity=supervisor_properties['num_groups'])
    
    # plt.figure(figsize=(10,10))
    # nx.draw(G, with_labels = True)
    # solve the weighted matching
    flow_dict = nx.min_cost_flow(G)
    print(flow_dict)
    for topic in all_topics:
        edges = flow_dict[topic]
        if edges['sink'] and not edges[topic+"_2"]:
            raise RuntimeError(
                'Solution with num_topics_used={n} is not valid.'.format(n=num_topics_used))

    # print(alltopiclist)
    # decode solution
    topic_assignments = {t: [] for t in all_topics}
    for student in students:
        edges = flow_dict[student]
        for target in edges:
            if edges[target]:
                topic_assignments[target].append(student)
                break

    supervisor_assignments = {s: [] for s in supervisors}
    for topic in all_topics:
        edges = flow_dict[topic+"_2"]
        for target in edges:
            if edges[target]:
                supervisor_assignments[target].append(topic)

    return topic_assignments, supervisor_assignments
```

Represent allocation in supervisor_assigments and topic_assignments dict

```python
num_students = len(students)
for n in range(1, num_students+1):
    try:
        topic_assignments, supervisor_assignments = get_student_topic_supervisor_assignments(all_topics, students, supervisors, num_topics_used=n)
        print(
            ' An optimal solution was found with `num_topics_used`={n}'.format(n=n))
        
        #Changing topic from ids to topic names in both dicts
        tp = {}
        for p in topic_assignments:
            tp[project_names[p]] = topic_assignments[p]
        topic_assignments = tp
        print(' Topic assignments:\n', topic_assignments)
        for column in supervisor_assignments:
            newlist = []
            for p in supervisor_assignments[column]:
                newlist.append(project_names[p])
            supervisor_assignments[column] = newlist

        print(' Supervisor assignments:\n', supervisor_assignments)
    except Exception as e:
        pass
```

Convert topic_assignments dict to topic_df dataframe

```python
lst_of_lst = []
for i,j in topic_assignments.items():
  row_lst = [i,j]
  lst_of_lst.append(row_lst)
print(lst_of_lst)
topic_df = pd.DataFrame(lst_of_lst, columns=['Topics', 'Students'])
display(topic_df)
```

Convert supervisor_assignments dict to topic_df dataframe

```python
lst_of_lst2 = []
for i,j in supervisor_assignments.items():
  row_lst = [i,j]
  lst_of_lst2.append(row_lst)
print(lst_of_lst2)
supervisor_df = pd.DataFrame(lst_of_lst2, columns=['Supervisor', 'Topics'])
display(supervisor_df)
```

Exploding topics column to split mutiple topics assigned to the same supervisor in different rows

```python
exploded = supervisor_df.explode('Topics')
display(exploded)
```

Join the two dataframes supervisor_df and topic_df

```python
df_output = pd.merge(exploded, topic_df, on='Topics')
```
